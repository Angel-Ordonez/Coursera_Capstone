```python
import pandas as pd

#Manera en que obtengo la data
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df_list = pd.read_html(url)

df = df_list [0]
df.head()
```

```python
#print(df.columns)
#df.columns = ['Postal Code', 'Municipality', 'Neighborhood']
df.columns = [ 'PostalCode' , 'Borough' , 'Neighborhood' , 'Latitude' , 'Longitude' , 'Country' , 'State' , 'County' , 'City' ]
#df = df_list [0].loc[:, ['PostalCode' , 'Neighborhood', 'Borough' ]]
df.head()
```

```python
#Recorrere los valores  de las columnas

code_list = df.values.tolist()
```

```python
#El objetivo es omitir las "Not assigned"

postal_code =[]
borough = []
nb = []
for i in code_list:
    for j in i:
        if j[3:] == 'Not assigned':
            pass
        else:
            postal_code.append(j[0:3])
            sep = j[3:].split('(')
            borough.append(sep[0])
            nb.append(sep[1].split(')')[0])
diccionario1 = {'Postal_Code':postal_code,'Borough':borough, 'Neighborhood':nb}
```

```python
df1 = pd.DataFrame(diccionario1)
df1.head(10)
```

```python
#Obtener los codigos postales para mandarselos a la api
PostalCodes = df1['Postal_Code'].tolist()
#PostalCodes
```

```python
#Aqui se va a tardar por va a usar la api en for, y son como 40 iteraciones que hara

latitudes = []
longitudes = []
codigopostales = []

for codigo in PostalCodes:
    api_key = "b23627b996694bb8a398fc5e1ff7eff7"
    geocoder = opencage.geocoder.OpenCageGeocode(api_key)
    query = codigo
    result = geocoder.geocode(query)
    if len(result) > 0:
        latitudes.append(result[0]['geometry']['lat'])
        longitudes.append(result[0]['geometry']['lng'])
        #codigopostales.append(result[0]["components"]["postcode"])
        codigopostales.append(codigo)
    diccionario2 = {'Postal_Code': codigopostales, 'Latitude': latitudes, 'Longitude': longitudes }
```

```python
df2 = pd.DataFrame(diccionario2)
df2.head()
```

```python
#DataFrame (Marco de datos) solicitado

df_completo = pd.merge(df1, df2, on='Postal_Code')
df_completo
```

```python
toronto_agrupado = df_completo.groupby('Neighborhood').mean().reset_index()
#toronto_agrupado.agg({"Cluster Label" : "mean"})
toronto_agrupado.head()
```

```python
toronto_agrupado.shape
```

```python
#Importar las librerias que necesitare para el grafico

from sklearn.cluster import KMeans
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

!conda install -c conda-forge folium=0.5.0 --yes
import folium 

!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim
```

```python
#Establecer el numero de agrupaciones

kclusters = 5
toronto_agrupado_clustering = toronto_agrupado.drop('Neighborhood', 1)

#ejecutar k-means
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_agrupado_clustering)

kmeans.labels_[0:10]                         #Esto lo insertare en el DataFrame de toronto_agrupado
```

```python
toronto_agrupado.insert(0, 'Cluster Labels', kmeans.labels_)
toronto_agrupado.head()
```

```python
#Necesite esto para obtener las latitudes y Longitudes de Toronto en el mapa
#Las latitudes y longitudes de toronto_agrupado van a ser los puntitos

address = 'Toronto'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address) 
latitude = location.latitude
longitude = location.longitude

print('La coordenas del mapa de Toronto son: ', latitude, ', ', longitude)
```

```python
#crear Mapa
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)


#Establecer el esquema de color para las agrupaciones
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

#aÃ±adir marcadores al mapa
#
markets_colors =[]
for lat, lon, poi, cluster in zip(toronto_agrupado['Latitude'], toronto_agrupado['Longitude'], toronto_agrupado['Neighborhood'], toronto_agrupado['Cluster Labels']):
    label = folium.Popup(str(poi) + 'cluster' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius = 5,
        popup = label,
        color = 'blue',              #rainbow, puse azul mejor porque se veian todo rojo, no se veia atractivo
        fill = True,
        fill_color = rainbow[cluster-1],
        fill_opacity = 0.7
    ).add_to(map_clusters)
    
map_clusters
```
